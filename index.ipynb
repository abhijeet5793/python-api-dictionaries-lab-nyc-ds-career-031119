### Q1: Load the first page of results with 50 results per page

```python
import math
import requests
import pandas as pd
import numpy as np
import json
page = 1
per_page = 50

url_params = { 'page' : page,
               'per_page' : per_page }

url = 'https://api.openbrewerydb.org/breweries?page=1&per_page=50'
# Make a get request to get the latest position of the international space station from the opennotify api.

response = requests.get("https://api.openbrewerydb.org/breweries?page=1&per_page=50", url_params)
# Print the status code of the response.

#print(response.text)
print(response.text)
```
